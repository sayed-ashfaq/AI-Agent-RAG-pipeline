```python
## Implementing from start
import os
from uuid import uuid4
from langchain_qdrant import QdrantVectorStore
from qdrant_client import QdrantClient
from qdrant_client.http.models import Distance, VectorParams
```

```python
from src.utils.model_loader import ModelLoader

embedding_model = ModelLoader().load_embeddings()
```

```python
# initiating client based on use case inmemory, local storage, cloud, server

# using in memory

client = QdrantClient(":memory:")
client.create_collection(
    collection_name="demo_collection",
    vectors_config=VectorParams(size= 3072, distance=Distance.COSINE)
)
```

```python
vector_store = QdrantVectorStore(
    client=client,
    embedding=embedding_model,
    collection_name="demo_collection"
)
```

```python
# create uuid for each document
uuids = [str(uuid4()) for _ in range(len(doc1))]

# add to the vectore store
vector_store.add_documents(documents=doc1, ids=uuids)
```

```python
retriever = vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 2})
result = retriever.invoke("What is langchain?")
result
```
